<center><h1>LSTM for k-hot Encoded Multiclassification:</h1></center>

```python
import numpy as np

import keras.backend as K
from keras.models import Sequential
from keras.layers import Dense, Activation, Reshape
from keras.layers import LSTM

from sklearn.cross_validation import train_test_split
```

<h4>Generate Random Training Lables:</h4>

```python
#Create the training labels
train_y = (np.random.randint(0,150,10000))
y = np.zeros((10000,150))
c = 0
for i in train_y: 
    y[c][i] = 1
    c = c+1
train_y = y.reshape(10000,1,150).reshape(10000,150)

#Random Multi encoded training labels
r = np.random.rand(10000,150)
r[r <= 0.3] = False
r[r > 0.3] = True
train_y[r == False] = 1
train_y = train_y.reshape(10000,1,150)
```

<h4>Generate Training X Data:</h4>

```python
#Create the training x data
train_x = np.array([np.random.rand(1, 1000)[0] for i in range(10000)]).reshape(10000,1,1000)
```

<h4>Train Test Split:</h4>

```python
train_x, test_x, train_y, test_y = train_test_split(train_x, train_y, test_size=0.2, random_state=42)
```

<h4>Create A Sample Query to Predict On:</h4>

```python
#Create the sample query 
sample = train_x[0].reshape(1,1,1000)
sample2 = train_x[12].reshape(1,1,1000)
```

<h4>Create Multiclass Loss Function:</h4>

```python
#Our custom loss function
def multiclass_loss(y_true, y_pred):
    EPS = 1e-5
    y_pred = K.clip(y_pred, EPS, 1 - EPS)
    return -K.mean((1 - y_true) * K.log(1 - y_pred) + y_true * K.log(y_pred))
```

<h4>Instantiate Parameters for LSTM:</h4>

```python
#Our inputs for our model
shape = train_x.shape[2]
print(shape)
num_classes = len(train_y[0][0])
```

<h4>Create and Compile LSTM:</h4>

```python
#Function to create and compile our model
def create_model(shape,num_classes):
    model = Sequential()
    model.add(LSTM(32, input_shape=(None,shape), return_sequences=True))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='sigmoid'))
    model.compile(loss=multiclass_loss, optimizer='adam', metrics=['mse'])
    return model
```

<h4>Train LSTM</h4>

```python
#Creating and training our model
model = create_model(shape,num_classes)
history = model.fit(train_x, train_y,
              batch_size=100, epochs=20,
              verbose = 1)
```

```python
loss, accuracy = model.evaluate(train_x, train_y, verbose=True)
print(loss)
print(accuracy)
model.predict_classes(sample)
model.predict_classes(sample2)
print(model.predict_classes(sample),model.predict(sample2))
```

<h4>Save LSTM Model</h4>

```python
from keras.models import load_model   #save_model requires h5py, need to pip install h5py
model.save('khot_LSTM.h5') 
```

<h4>Load LSTM With Custom Loss Function:</h4>

```python
#Loading our model with our custom loss function
model = load_model('khot_LSTM.h5', custom_objects={"multiclass_loss":multiclass_loss})
```

<h4>Obtain Prediction of Sample Query:</h4>

```python
pred = model.predict(sample)[0][0]
#print(pred)
print('Our top prediction is: ',np.argmax(pred),"\n")
# or we can do 
print('Our top prediction is: ', model.predict_classes(sample)[0][0],"\n")
print("Indices of Top 3 predictions:")
top3 = pred.argsort()[-3:][::-1]
print(top3,"\n")
print("Indices where we have a 1:")
print(np.argwhere(train_y[0] == 1))
#print(pred)
#print(min(pred))
#print(max(pred))
#print(np.mean(pred))
```

<h4>Real Example:</h4>

```python
import pandas_datareader.data as web

start = '2008-01-01'
end = '2017-01-01'
df_AAPL = web.DataReader('AAPL',"yahoo",start,end)

# GET LABELS (if next price was greater then 1 else lowe is 0)
dollar_change = df_AAPL["Adj Close"] - df_AAPL["Adj Close"].shift(-1)
dollar_change = dollar_change.apply(lambda i: 0 if float(i) >= 0 else 1 )
labels = dollar_change.values
labels = labels[0:-1]
```

```python
#Normalize Features
df_AAPL_norm =(df_AAPL-df_AAPL.mean())/df_AAPL.std()

#Drop last row because of NaN
df_AAPL_norm = df_AAPL_norm[0:-1]
```

```python
x_features = df_AAPL_norm.values
```

```python
x = np.zeros((len(x_features),2))
c = 0
for i in labels:
    
    #This means buy: [0,1]
    if i == 1:
        x[c][1] = 1
        
    #This means sell [1,0]
    else: 
        x[c][0] = 1
    c = c +1 
labels = x
```

```python
#reshaping for LSTM 
num_samples = len(x_features)
num_classes = len(labels[0])
num_features = len(x_features[0])

#reshape labels
labels = labels.reshape(num_samples,1,num_classes)

#reshape features
x_features = x_features.reshape(num_samples,1,num_features)
```

```python
shape = x_features.shape[2]
print(shape)
num_classes = len(labels[0][0])
print(num_classes)
```

```python
def create_model(shape,num_classes):
    model = Sequential()
    model.add(LSTM(32, input_shape=(None,shape), return_sequences=True))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    model.compile(loss="binary_crossentropy", optimizer='adam', metrics=['mse'])
    return model
```

```python
model = create_model(shape,num_classes)
history = model.fit(x_features, labels,
              batch_size=100, epochs=200,
              verbose = 1)
```
